```python
import pandas as pd
import numpy as np
```

```python
true=pd.read_csv('True.csv')
fake=pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true['label']=1
fake['label']=0
```

```python
true.head()
```

```python
news=pd.concat([fake,true],axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news=news.drop(['title','subject','date'],axis=1)
```

```python
news.head()
```

```python
news=news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[^\w\s]','',text)
    text=re.sub('\n','',text)
    text=re.sub('\d','',text)

    return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizaion = TfidfVectorizer()
xv_train = vectorizaion.fit_transform(x_train)
xv_test = vectorizaion.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
LR=LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC=DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc=DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
predict_rfc=rfc.fit(xv_train,y_train)
```

```python
predicy_rfc=rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predicy_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier()
```

```python
gbc.fit(xv_train,y_train)
```

```python
print(classification_report(y_test,gbc.predict(xv_test)))
```

```python
def output_label(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "True News"
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorizaion.transform(new_x_test)

    # Assuming X_train and y_train are available globally
    # If not, you'll need to load or create them within this function
    global xv_train
    global y_train

    LR.fit(xv_train, y_train)
    pred_LR = LR.predict(new_xv_test)
    pred_DTC = DTC.predict(new_xv_test)
    pred_RFC = rfc.predict(new_xv_test)
    pred_GBC = gbc.predict(new_xv_test)
    return print("\n\nLR Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {} \nGBC Prediction: {}".format(output_label(pred_LR[0]), output_label(pred_DTC[0]), output_label(pred_RFC[0]), output_label(pred_GBC[0])))
```

```python
news_article=str(input())
manual_testing(news_article)
```
